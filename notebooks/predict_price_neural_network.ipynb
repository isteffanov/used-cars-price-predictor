```python
import pandas as pd
```

```python
from price_predictor.prepare_data import get_numerical, get_numerical_and_boolean, split_train_test_valid
from price_predictor.evaluate import evaluate_model, evaluate
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
import tensorflow.keras as tf
import tensorflow.keras.layers
```

```python
data_all = pd.read_csv('data/cars.csv')
data_all
```

```python
data_all.dropna(inplace=True)
```

```python
data = get_numerical(data_all)
data
```

```python
scaler = MinMaxScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
```

```python
data_scaled
```

```python
X_train, X_test, X_valid, y_train, y_test, y_valid = split_train_test_valid(data_scaled)
```

```python
len(X_train.columns)
```

```python
model = tf.Sequential()
model.add(tf.layers.Dense(len(X_train)))
model.add(tf.layers.Dense(256, activation='relu'))
model.add(tf.layers.Dense(256, activation='relu'))
model.add(tf.layers.Dense(1, activation='linear'))
```

```python
model.compile(loss='mean_absolute_error', optimizer='adam')
```

```python
model.fit(X_train, y_train, batch_size = 32, epochs=10)
```

```python
evaluate_model(model, X_test, y_test)
```

```python
evaluate_model(model, X_train, y_train)
```
